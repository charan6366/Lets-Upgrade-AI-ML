Statistical Test : Parametric and Non Parametric Test
Non Parametric test:
Data is not normally distributed..skewness is not zero.Hence, we willl go for Non-parametric test.
Since,we have 2 independent sample we can go for Mann-Whitney Test.
We can also go for Chi-Square test because there are categorical variables also.
Employees with Attrition Yes = sample1
Employess with Attrition No = sample 2

```python
import pandas as pd
df=pd.read_csv("general_data.csv")
df.head()
```

```python
dummies=pd.get_dummies(df1.Attrition,prefix="Attrition")
dummies
```

```python
merged=pd.concat([df1,dummies],axis="columns")
merged.head()
```

```python
final=merged.drop(["Attrition","Attrition_No"],axis="columns")
final.head()
```

Mann - Whitney test:

Hypothesis 1: Age
H0 = There is no significance difference in the age between the employees who stayed and the employees who left.
H1 = There is significance difference in the age between the employees who stayed and the employees who left.

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(empl_stayed.Age,empl_left.Age)
print(stats,p)
if p<0.05:
    print("H1 Accepted")
else:
    print("H0 Accepted")
```

Hypothesis 1 : Attrition and BusinessTravel
H0 = There is no dependency between Attrition and Business Travel
H1 = There is dependency between Attrition and Business Travel

```python
from scipy.stats import chi2_contingency
chitable=pd.crosstab(final.Attrition_Yes,final.BusinessTravel)
chitable
```

```python
stats,p,df,expected = chi2_contingency(chitable)
print(stats,p)
if p<0.05:
    print("H1 Accepted")
else:
    print("H0 Accepted")
```
