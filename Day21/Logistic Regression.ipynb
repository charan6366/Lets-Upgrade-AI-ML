```python
import statsmodels.api as sm
import math as m
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel(r"C:\Users\ADMIN\Desktop\LetsUpgrade\Datasets\Bank_Personal_Loan_Modelling.xlsx", sheet_name=1)
data.head()
```

```python
data = data.drop(columns = ['ID', 'ZIP Code'] )
data.drop_duplicates(inplace=True)
data.shape
```

```python
Y = data['Personal Loan']
X = data[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Securities Account', 'CD Account', 'Online', 'CreditCard']]
```

```python
X1 = sm.add_constant(X)
Logistic = sm.Logit(Y,X1)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```

# Inferences:-
When a bank wants to offer a loan, the following factors are highly significant:

Annual income of the customer
1.Family size of the customer

2.Education Level. 1: Undergrad; 2: Graduate; 3: Advanced/Professional

3.Does the customer have a certificate of deposit (CD) account with the bank?

4.Does the customer use internet banking facilities?

5.Does the customer use a credit card issued by UniversalBank?

6.Avg. spending on credit cards per month.

7.Does the customer have a securities account with the bank?

```python
dataset = pd.read_csv(r"C:\Users\ADMIN\Desktop\LetsUpgrade\Datasets\general_data.csv")
dataset.head()
```

```python
from sklearn import preprocessing
a = preprocessing.LabelEncoder()
dataset['Attrition'] = a.fit_transform(dataset['Attrition'])
dataset['BusinessTravel'] = a.fit_transform(dataset['BusinessTravel'])
dataset['EducationField'] = a.fit_transform(dataset['EducationField'])
dataset['Gender'] = a.fit_transform(dataset['Gender'])
dataset['Department'] = a.fit_transform(dataset['Department'])
dataset['MaritalStatus'] = a.fit_transform(dataset['MaritalStatus'])
dataset['JobRole'] = a.fit_transform(dataset['JobRole'])
```

```python
Y1 = dataset['Attrition']
X2 = dataset.drop(columns = ['Attrition'])
```

```python
X3 = sm.add_constant(X2)
Logistic1 = sm.Logit(Y1,X3)
```

```python
res = Logistic1.fit()
```

```python
res.summary()
```

The following factors highly affect attrition in employees:

Age of the employee

Marital status of the employee

Total number of companies the employee has worked for
